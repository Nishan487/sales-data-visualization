```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

sales_dataset = pd.read_excel("SaleData.xlsx")
```

```python
sales_dataset
```

```python
avg_sales=sales_dataset['Sale_amt'].mean()
```

```python
avg_sales
```

```python
max_order=sales_dataset['Units'].max()
print(max_order)
```

```python
item_detail=sales_dataset.groupby("Item")["Sale_amt"].mean()
item_detail.plot(kind="bar",color="orange",figsize=(10,5))
plt.title("Average Sales Amount per item")
plt.ylabel("Average Sales (Rs)")
plt.xlabel("Food Item")
plt.show()
```
